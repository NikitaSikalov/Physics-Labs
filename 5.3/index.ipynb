## Определение энергии $\gamma$-квантов с помощью сцинтилляционного спектрометра

### Цель работы: Определение энергии $\gamma$-квантов неизвестного радиоактивного препарата после предварительной калибровки спектрометра по $\gamma$-излучению ${^{60}}Co$.

### I. Теория

При фотоэффекте $\gamma$-квант целиком поглощается атомом, а один электронво внутрнней оболочки выбрасывается за пределы атома и уносит всю переданную энергию $\gamma$-квантом энергию. Энергию конверсионного электрона: $T_e=E_{\gamma}-W_K$, где $W_K$ - энергия ионизации для K - оболочки атома. При фотоэффекте в кристалле выделяется энергия, практически равная энергии поглощенного $\gamma$-кванта. Амплитуда световых вспышек ~ полной энергии первичных $\gamma$-квантов. 

Комптоновское рассеяние $\gamma$-квантов происходит на слабо связанных электронах. Наименьшая переданная энергия электрону - 0, наибольшая: $$(T_e)_{max} = \hbar \omega \frac{2 \alpha}{1 + 2 \alpha},$$ где $\alpha = \hbar \omega / (mc^2).$ Рассеянный $\gamma$-квант может либо поглотиться, либо покинуть кристалл. 

Если энергия $\gamma$-квантов превышает величину $2mc^2=1.02$МэВ, то становится возможным образование пары электрон-позитрон в поле ядра. Кинитическая энергия электрона и позитрона равна: $$T_{e^{-}} + T_{e^{+}} = \hbar \omega - 2mc^2.$$
Оставшийся позитрон аннигилирует с одним из атомных электронов $\gamma$-квантам. Оставшаяся в кристалле энергия может поэтому равняться  энергии $\gamma$-кванта или отличаться от нее на $mc^2$ или $2mc^2$

Интегральные кривые распределения импульсов представляют собой зависимость числа имапульсов $N(E)$. Ниже приведены примеры таких импульсов.

### II. Эксперементальная установка
---
Принцип действия: $\gamma$-кванты от исследуемого источника коллимируются йодистого натрия, активированного таллием, NaI(Tl). Попавшие в сцинтиллятор кванты взаимодействуют с его атомами, выбивая электроны из атомных оболочек и при достаточно большой энергии генерируя электрон-позитронные пары. Вырванные из атомных оболочек электроны (конверсионные) тормозятся в веществе сцинтиллятора, расходая энергию на возбуждение атомов среды. Возбужденные атомы высвечиваются , испуская ЭМ излучение. Образовавшиеся фотоны, пройдя через сцинтиллятор, попадают на катод фотоэлектроного умножителя и вырывают с поверхности первого динода вторичные электроны, которые ускоряются полем. Возникшие на аноде ФЭУ импульсы фиксируются компьютером.

```python
import pandas as pd
import matplotlib.pyplot as plt

font={'fontsize': '20'}
table=pd.read_excel("Co_60.xlsx")
Channel = table.values[:, 0]
Counts = table.values[:, 1]
plt.figure(figsize=(15, 8))
plt.title("Спектрометрия ${^{60}}Co$", **font)
plt.ylabel("Counts", size=20)
plt.xlabel("Channel", **font)
plt.plot(Channel, Counts)
plt.grid(True)
plt.show()
```
