### I. Теория
При $\alpha$-распаде атома испускаютя ядра ${}^4_2He$. Для ${}^{212}_{84}Po$ и ${}^{204}_{82}Pb$ энергия $\alpha$-частиц лежит в пределах от 4 до 9 _МэВ_. Связь между периодом полураспада и энергией частиц описывается формулой: $$lgT_{1/2} = \cfrac{a}{\sqrt E} + b. \tag{1}$$
Эксперементальную энергию $\alpha$-частиц удобно определять по величине их пробега в веществе. Неупругие столкновения вызывют ионизацию атомов и небольшие потери энергии у заряженной частицы. Потери --- ионизационные. Энергия отдаваемая электрону не превышает $4mE/M$. При этом $\alpha$-частица отклоняется на очень малый угол (не более чем $\frac{m}{M}$). Заметим, что энергия частиц ниже кулоновского барьера, поэтому ядерные реакции практически невозможны. 

Рассмотрим теорию о расчете удельных потреь энергии заряженной частицы в веществе. Пусть частица с сзарядом $z$, движущаяся в направлении $x,$ проходит на расстоянии $y$ от покоящегося свободного электрона. В перпендикулярном направлении движения частицы перадается импульс силы $\left( Ze^2 / y^2 \right)$ за время $\approx 2y / v$ Потерянная энергия частицы будет определяться выражением: $$E_e = \cfrac{p^2}{2m} = \cfrac{1}{2m} \left( \cfrac{Ze^2}{y^2} \cdot \cfrac{2y}{v} \right) ^2 = \cfrac{2e^4Z^2}{mv^2y^2}. \tag{2}$$
В слое $2\pi y dy$ $\alpha$-частиц потеряет $$dE(y) = \cfrac{4\pi n Z z^2 e^4}{mv^2}\cfrac{dy}{y}. \tag{3}$$ 
Полная потеря энергии составит $$\left( \cfrac{dE}{dx} \right)_{ион} \approx 4\pi \cfrac{e^4 z^2}{mv^2}nZ \cdot ln\cfrac{y_{max}}{y_{min}}. \tag{4}$$
Из формулы (2) следует, что энергия потерянная заряженной частицей при столкновении обратно пропорцианальна квадрату прицельного парметра, т. е. $$2ln\cfrac{y_{max}}{y_{min}} = -ln\cfrac{E_{max}}{E_{min}}. \tag{5}$$
Из закона сохранении энергии следует: $$E_{max} = \cfrac{4mE}{M} = 2mv^2. \tag{6}$$ 
Минимальная энергия, передаваемая электрону, определяется энергией возбуждения или энергии связи электрона. Оно характеризуется _средним иониазационным потенциалом $\overline{I}$_ $$ln\cfrac{E_{max}}{E_{min}} = ln\cfrac{2mv^2}{\overline{I}}, \tag{7}$$
Тогда в нерелятивистском случае получим формулу для _тормнозной способности вещества_: 
$$\left( \cfrac{dE}{dx} \right)_{ион} \approx 2\pi \cfrac{e^4z^2}{mv^2}nZ \cdot ln \cfrac{2mv^2}{\overline{I}}. \tag{8}$$
Отсюда мы можем вычислить длину пробега частицы в следующем виде: 
$$R_{zM} = -\int\limits_{E_1}^{E_0}\cfrac{dE}{(dE/dx)} = \cfrac{m}{2\pi e^4Z^2nZ}\int\limits_{E_1}^{E_0}\cfrac{v^2dE}{ln(2mv^2/\overline{I}}. \tag{9}$$
Принимая во внимание, что $dE = Mvdv$, получаем 
$$R_{zM} = \cfrac{mM}{2\pi e^4z^2nZ}\int\limits_{v_1}^{v_0}\cfrac{v^3dv}{ln(2mv^2/\overline{I})}. \tag{10}$$
Если принеберечь слабой логарифмической зависимостью от скорости частицы, то
$$R \propto \cfrac{M}{z^2}v_0^4 \propto E^2. \tag{11}$$
От 4 до 9 _МэВ_ эксперимент хорошо согласуется с формулой:
$$R = 0,32E^{3/2}. \tag{12}$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as sps
```

```python
N = np.array([104 / 7.778, 106 / 7.283, 103 / 6.678, 105 / 8.069, 12 / 8.692, 12 / 21.654, 9 / 25.341, 6 / 15.532, 5 / 35.241])
x = np.array([0, 3, 5, 7, 9, 10, 12, 15, 20])
grid = np.linspace(0, 20, 100)
plt.figure(figsize=(12, 10))
plt.title('График зависимсоти частоты фиксирования импульсов от удаленности источника', fontsize=20)
plt.yticks(np.arange(0., 20., 1.))
plt.xticks(np.arange(-1., 21., 1.))
plt.ylim(-0.5, 17)
plt.ylabel(r'$N_{\tau}, \: Гц$', fontsize=15)
plt.xlabel('x, мм', fontsize=15)
plt.grid(which='both')
plt.plot(x, N, 'o')
f = sps.interp1d(x, N, kind='quadratic')
h = 0.01
x0 = 7.8
k = (f(x0 + h) - f(x0)) / h
tan = f(x0) + k * (grid - x0) 
plt.plot(grid, f(grid), 'r')
plt.plot(grid, tan, 'k--')
plt.plot(np.ones(10) * x0, np.linspace(-100, 100., 10), 'c--')
```

Таким образом из графика видим, что $R_{ср} = 7,8$ мм и $R_э = 9,0$ мм. Если выражать пробег в $г/см^2$, то $R' = R\rho$. Считаем что измерения проводились при н. у., тогда $\rho = 1,225 кг/м^3$, тогда $R'_{ср} = 9.56 \: г/см^3$, $R_э = 11.03 \: г/см^3.$ ПО формуле (12) определим $E$ - энергия частицы: $E_{\alpha} = 2$ _МэВ_

Из графика находим эксраполированный пробег $\alpha$-частицы как точку пересечения касательных: $P_0 = 555 \: торр$.
