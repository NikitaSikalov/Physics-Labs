### I.Теория

Электроны эмитируемые катодом тиратрона, ускоряются напряжением $V$, приложенным между катодом и сеткой. Затем электроны рассеиваются на атомах инертного газа. Все сетки соеденены между собой и имеют одинаковый потенциал, примерно равный потенциалу анода. Рассеянные электроны отклоняются и уходят на сетку, а оставшаяся часть электронов достигает анода и создает аноднй ток $I_a$. Таким образом, поток электронов в $N(x)$ на расстоянии $x$ от ускоряющей сетки уменьшается с ростом $x$ от начального значения $N_0$ у катода до некоторого значения $N_a$ у анода.

Запишем убыль электронов в результате прохождения слоя $dx$:
$$-\frac{dN}{N(x)} = \frac{\Delta}{S}\omega(V) = n_a \Delta_a \omega(V) dx, \tag{1}$$ где $\Delta_a$ -площадь поперечного сечения атома, а $\omega(V)$ - вероятность рассеяния на атоме. Интегрируя это выражение от $0$ до $L$ и заменяя поток электронов на ток $I = Ne$, получаем уравнение ВАХ: 
$$I_a = I_0 e^{-C\omega(V)}, \; C = Ln_a\Delta_a, \tag{2}$$ где $I_0 = eN_0$, $I_a = eN_a$.

Отсюда выводим вероятность рассеяния электрона от его энергии соотношением:
$$\omega(V) = -\frac{1}{C}ln\frac{I_a}{I_0}. \tag{3}$$

Также зная энергию электронов можно определить эффективный размер атома: 
$$2l = \frac{3}{2}\frac{h}{\sqrt{2m(E_2 + U_0)}}, \tag{4}$$
избавлясь от $U_0$ получим: 
$$l = \frac{h\sqrt{5}}{\sqrt{32m(E_2 - E_1)}} \tag{5}$$
Из этих формул получаем эффективную глубину ямы атома: 
$$U_0 = \frac{4}{5}E_2 - \frac{9}{5}E_1. \tag{6}$$

###  III. Выполнение

$\Delta V = 6В$
$V_{накала} = 2В$

Отсюда $E_1 = 2 \,эВ$, $E_2 = 8 \, эВ$, $U_0 = 2.5эВ$

Посчитаем размер электронной оболочки по формуле (4): $l = 1.6\, A$. 

Теперь посчитаем то же значение по формуле (5) не зная $U_0$: $l = 3 \, A$

Рассчитаем глубину потенциальной ямы по формуле (6): $U_0 = 1.8 \, эВ$

Расчитаем ионизационный потенциал Аргона: $E_и = 15.4 \, эВ \approx 15.8 \, эВ$, то есть тиратрон наполнен аргоном.

```python
%matplotlib inline 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline
plt.rcParams.update({'font.size': 22, 'figure.figsize': [12, 10], 'axes.grid': True, 'lines.markersize': 20, 'lines.markeredgewidth': 3})
```

```python
data1 = pd.read_csv("Накал_2_60.csv")
data1
```

```python
X1 = data1.values[:,0]
Y1 = data1.values[:,1]
grid = np.linspace(X1.min(), X1.max())
f = CubicSpline(X1, Y1)
plt.figure(figsize=(14, 10))
plt.title("Ток анода от напряжения катод-сетка")
plt.plot(X1, Y1, "+")
plt.plot(grid, f(grid), label="$V_н = 2.60 В$")
plt.ylabel(r"$V_a, мВ$")
plt.xlabel(r"$V_к$, В")
plt.legend()
```

```python
data2 = pd.read_csv("Накал_1_7.csv")
data2
```

```python
X2 = data2.values[:,0]
Y2 = data2.values[:,1]
grid = np.linspace(X2.min(), X2.max())
f = CubicSpline(X2, Y2)
plt.figure(figsize=(14, 10))
plt.title("Ток анода от напряжения катод-сетка")
plt.plot(X2, Y2, "+")
plt.plot(grid, f(grid), label="$V_н = 1.7 В$")
plt.ylabel(r"$V_a, мВ$")
plt.xlabel(r"$V_к$, В")
plt.legend()
```

Найдем зависимость $E_n = f(E_1, n) = n^2(E_1 + U_0) - U_0$. Отсюда получаем, следующие значения $E_2 = 13,4 \, эВ$, $E_3 = 32,4 \, эВ$. Построим график зависимсоти вероятности рассеяния электронов от энергии:

```python
C = 5
omega = -(1/C) * np.log(Y1 / X1) + 1
grid = np.linspace(X1.min(), X1.max())
f = CubicSpline(X1, omega)
plt.figure(figsize=(14, 10))
plt.title("Зависимость вероятности рассеяния электронов от энергии")
plt.plot(X1, omega, "+")
plt.plot(grid, f(grid), label="$V_н = 2.6 В$")
plt.xlabel(r"$V_к, В$")
plt.ylabel(r"$\omega(V)$")
plt.legend()
```
