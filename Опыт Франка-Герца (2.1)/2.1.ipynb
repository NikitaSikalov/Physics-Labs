### Цель работы: методом эдектронного возюуждения измеряется энергия первого уровня атома гелия в динамических и статистических режимах

### I. Теория
**Описание опыта**

Разреженный одноатомный газ заполняет трехэлектроднуб лампу. Электроны ускоряются полем созданным между катодом и сетчатым анодом. Передвигаясь они сталкиваются с атомами гелия. Если энергия электрона не очеь большая, то соударения будут упругими и электрон практически не попотеряет энергии. Однако, если энергии электрона будет достаточно, чтобы перевести атом в первое возбужденное состояние, то из-за неупругого столкновения он может потерять почти всю свою энергию и переведя атом в возбужденное состояние. 
<img src="Selection_065.png" width=650>
Третьим электродом является коллектор, мужду ним и анадом поддерживает небольшое задерживающее напряжение. При увелечении потенциала анода ток в лампе вначале растет, однако когда энергия электронов становится достаточной для возбуждения атомов, ток коллектора резко уменьшается. Таким образом, на кривой зависимости тока коллектра от напряжения анода имеется ряд максимумов и минимумумов, отстоящих друг от друга на равные расстояния $\Delta V$; эти расстояния равны энергии первого возбужденного состояния.

### II. Экспериментальная установка и выполнение работы

В качестве анода используется двойная спираль, окружающая катод. Роль коллектора играет полый металлический цилиндр, соосный с катодом и анодом. Источником электронов является вольфрамовый катод, нагреваемый переменным током. Ток в цепи коллектора рагестрируется потенциометром $П_2$ и измеряется вольтметром $V_2$.

### III. Выполнение работы
1. Измерим расстояние между максимумами на на осциллографе:

| $V_0$, В | $\Delta_1$, В | $\Delta_2$, В |
| :---: | :--------: | :--------: |
| 4 | 15.0 | 18.5 |
| 6 | 14.0  |  17.0 |
|8 | 8.5 | 13.5 |
По этим предворительным измерения можно считать, что энергия 1-го урвоня гелия $E_1 \approx 17$ эВ.

2. Снимем ВАХ $I_к = f(V_a)$:

```python
%matplotlib inline 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline
plt.rcParams.update({'font.size': 22, 'figure.figsize': [12, 10], 'axes.grid': True, 'lines.markersize': 20, 'lines.markeredgewidth': 3})
```

```python
files = np.array(["4.csv", "6.csv", "8.csv"])
params = np.array([4, 6, 8])
i = 0
for file in files:
    data = pd.read_csv(file)
    print("\nЗадерживающее напряжение - " + str(params[i]) + 'В')
    print(data)
    I = data.values[:,1]
    V = data.values[:,0]
    plt.plot(V, I, '+')
    cs = CubicSpline(V, I, bc_type=((2, 0), (2, 0)))
    grid = np.linspace(0, 90)
    plt.plot(grid, cs(grid), label=str(params[i]) + "В")
    i += 1
plt.title(r"ВАХ тока катода от напряжения между анадом и катодом")
plt.xlabel(r"$V_а$")
plt.ylabel(r"$I_к$")

plt.legend()
```

Из графиков видно, что рассстояние между макисимумами $\Delta V \approx 18$ В, отюсда энергия первого возбужденного уровня $He$ состовляет около $E_ 1 \approx 18$ эВ. Из справочника $E_1 = 21.2\; \text{эВ}$
